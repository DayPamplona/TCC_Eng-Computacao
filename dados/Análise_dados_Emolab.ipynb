```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ----------- Simulando um dataset (substitua pela sua coluna de interesse) -----------
# Exemplo de geração de dados aleatórios
# Se você quiser rodar no seu dataset, substitua a linha abaixo pelo seu DataFrame e coluna
import numpy as np
np.random.seed(42)
dados = np.random.normal(loc=5000, scale=1500, size=1000)  # Média 5000, desvio 1500

# Convertendo para Series
serie = pd.Series(dados)

# ----------- Cálculo das estatísticas -----------
media = serie.mean()
mediana = serie.median()
desvio = serie.std()
variancia = serie.var()
maximo = serie.max()
minimo = serie.min()
amplitude = maximo - minimo

# ----------- Impressão dos resultados -----------
print(f"Média: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Desvio Padrão: {desvio:.2f}")
print(f"Variância: {variancia:.2f}")
print(f"Máximo: {maximo:.2f}")
print(f"Mínimo: {minimo:.2f}")
print(f"Amplitude: {amplitude:.2f}")

# ----------- Plotando o gráfico -----------
plt.figure(figsize=(12,6))
sns.histplot(serie, kde=True, bins=30, color="lightblue")

# Linhas da média e mediana
plt.axvline(media, color='green', linestyle='-', linewidth=2, label=f'Média: {media:.2f}')
plt.axvline(mediana, color='blue', linestyle='-', linewidth=2, label=f'Mediana: {mediana:.2f}')

# Linhas de desvio padrão (positivo e negativo)
plt.axvline(media + desvio, color='orange', linestyle='--', linewidth=2, label=f'+1 Desvio: {media + desvio:.2f}')
plt.axvline(media - desvio, color='orange', linestyle='--', linewidth=2, label=f'-1 Desvio: {media - desvio:.2f}')

# Título e legendas
plt.title('Distribuição dos Dados com Média, Mediana e Desvio Padrão')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.legend()

plt.show()
```

```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 🔗 Leitura do CSV
from google.colab import files
uploaded = files.upload()

df = pd.read_csv(list(uploaded.keys())[0])

# Verificar os dados
print(df)

# ------------------ 🔵 Gráfico de Attention ------------------

plt.figure(figsize=(10,6))

# Plot das barras
plt.bar(df['voluntario'], df['attention_mean'],
        yerr=df['attention_std'], color='skyblue', capsize=8)

# Linhas horizontais tracejadas
max_y = (df['attention_mean'] + df['attention_std']).max() * 1.2
plt.yticks(np.arange(0, max_y, step=10))
for y in np.arange(0, max_y, 10):
    plt.axhline(y=y, color='gray', linestyle='--', linewidth=0.5)

# Customização do gráfico
plt.title('Média de Attention por Voluntário (com Desvio Padrão)')
plt.ylabel('Attention')
plt.xlabel('Voluntário')
plt.xticks(rotation=45)
plt.ylim(bottom=None)  # Sem limite inferior fixo
plt.tight_layout()

plt.show()

# ------------------ 🟢 Gráfico de Engagement ------------------

plt.figure(figsize=(10,6))

# Plot das barras COM desvio padrão completo (permitindo ir para negativo)
plt.bar(df['voluntario'], df['engagement_mean'],
        yerr=df['engagement_std'],
        color='lightgreen', capsize=8)

# Linhas horizontais tracejadas
max_y = (df['engagement_mean'] + df['engagement_std']).max() * 1.2
min_y = (df['engagement_mean'] - df['engagement_std']).min() * 1.2

# Se houver valores negativos, o eixo Y será ajustado automaticamente
plt.yticks(np.arange(int(min(min_y, 0)), int(max_y)+10, step=10))
for y in np.arange(int(min(min_y, 0)), int(max_y)+10, 10):
    plt.axhline(y=y, color='gray', linestyle='--', linewidth=0.5)

# Customização do gráfico
plt.title('Média de Engagement por Voluntário (com Desvio Padrão)')
plt.ylabel('Engagement')
plt.xlabel('Voluntário')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 🔗 Leitura do CSV
from google.colab import files
uploaded = files.upload()

df = pd.read_csv(list(uploaded.keys())[0])

# Verificar os dados
print(df)

# ------------------ 🎯 Gráfico agrupado (Attention e Engagement) ------------------

# Número de voluntários
n = len(df['voluntario'])

# Posições das barras no eixo X
x = np.arange(n)

# Largura de cada barra
width = 0.35

# Criar a figura
plt.figure(figsize=(12,6))

# Plotar as barras de Attention
plt.bar(x - width/2, df['attention_mean'],
        width,
        yerr=df['attention_std'],
        capsize=8,
        label='Atenção',
        color='skyblue')

# Plotar as barras de Engagement
plt.bar(x + width/2, df['engagement_mean'],
        width,
        yerr=df['engagement_std'],
        capsize=8,
        label='Engajamento',
        color='lightgreen')

# Definir labels no eixo X
plt.xticks(x, df['voluntario'], rotation=45)

# Linhas horizontais tracejadas
max_y = (max(df['attention_mean'] + df['attention_std'].fillna(0)) +
         max(df['engagement_mean'] + df['engagement_std'].fillna(0))) * 0.6

plt.yticks(np.arange(0, max_y+20, step=10))
for y in np.arange(0, max_y+20, 10):
    plt.axhline(y=y, color='gray', linestyle='--', linewidth=0.5)

# Título e legendas
plt.title('Média de Atenção e Engajamento por Voluntário - Atividade Nível Fácil')
plt.ylabel('Média')
plt.xlabel('Voluntário')
plt.legend()

plt.tight_layout()
plt.show()
```
